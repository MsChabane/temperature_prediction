```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df =pd.read_csv('../data/cleaned_data.csv')
```

```python
df.precip_type=df.precip_type.fillna("messing")
```

```python
num_cols=df.columns[2:].tolist()
cat_cols=df.columns[:2].tolist()
```

```python
sns.countplot(df,y=cat_cols[0]);
```

```python
sns.countplot(df,y=cat_cols[1]);
```

```python
for col in num_cols:
    fig,ax=plt.subplots(1,2,figsize=(10,3))
    sns.kdeplot(df,x=col,fill=True,ax=ax[0])
    sns.boxplot(df,x=col,fill=True,ax=ax[1])
    plt.tight_layout()
    plt.show()
```

```python
sns.pairplot(df[num_cols])
```

```python
corr=df[num_cols].corr()
mask=np.triu(np.ones_like(corr))
sns.heatmap(corr,mask=mask,cmap='Blues',annot=True,fmt=".2f");
```
